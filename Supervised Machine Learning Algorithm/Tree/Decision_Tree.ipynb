```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
from sklearn.datasets import load_iris
#Classification:
iris=load_iris()
```

```python
print(iris["DESCR"])
```

```python
#Independent Variable:
X=pd.DataFrame(iris['data'],columns=['sepal length','sepal width','petal length','petal width'])
X
```

```python
#Dependent Variable:
Y=iris['target']
```

```python
#Train-Test-Split
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=10,test_size=0.2)
```

```python
classifier=DecisionTreeClassifier()
```

```python
classifier.fit(X_train,Y_train)
```

```python
Y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(Y_test,Y_pred)
```

```python
print(f"Accuracy is {acc}")
```

```python
plt.figure(figsize=(12,8))
plot_tree(classifier,filled=True)

#plt.savefig("DT.png")
```

Each Node consists of the feature in which the split is made, gini impurity(measure node purity), total samples, and the samples in each class. So, for the root node the splitting is taking place on the 3rd feature which is the petal length. The first split is a pure split in which all the dataset belongs to Iris-Setosa class. So, basically if the petal length is less than or equal to 2.45cm it belongs to Iris-Setosa category. Any plant with a petal length greater than 2.45cm can be either Iris-Versicolour or Iris-Virginica. To differntiate, it again makes another split based on petal length and it goes on till level 6 where it makes a completely pure split.

Here, we can see the obvious flaw of Decision tree, that is since there is no stopping criteria, it tends to overfit and create pure nodes.

```python
from sklearn.datasets import load_diabetes
```

```python
#Regression:
diabetes=load_diabetes()
```

```python
print(diabetes['DESCR'])
```

```python
X2=pd.DataFrame(diabetes.data,columns=['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
Y2=diabetes['target']
```

```python
x_train,x_test,y_train,y_test=train_test_split(X2,Y2,test_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
clf=DecisionTreeRegressor()
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import numpy as np

print(f"mean absolute error: {mean_absolute_error(y_test,y_pred)}")
print(f"mean squared error: {mean_squared_error(y_test,y_pred)}")
print(f"root mean square error: {np.sqrt(mean_squared_error(y_test,y_pred))}")
```

```python
#Hyperparameter tuning:
params={
    'criterion':['squared_error','friedman_mse','absolute_error','poisson'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,6,35,20],
    'max_features':['auto','sqrt','log2']   
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid=GridSearchCV(clf,param_grid=params,cv=5,scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')

grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(x_test)
```

```python
print(f"mean absolute error: {mean_absolute_error(y_test,y_pred)}")
print(f"mean squared error: {mean_squared_error(y_test,y_pred)}")
print(f"root mean square error: {np.sqrt(mean_squared_error(y_test,y_pred))}")
```
